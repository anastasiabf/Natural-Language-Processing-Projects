```python
# Converting the 'placeID' data series into a list
resto_id = preparation['placeID'].tolist()

# Converting the 'Name' data series into a list
resto_name = preparation['name'].tolist()

# Converting the 'Rcuisine' data series into a list
resto_cuisine = preparation['Rcuisine'].tolist()

# Printing the lengths of the converted lists
print(len(resto_id))
print(len(resto_name))
print(len(resto_cuisine))
```

```python
# Creating a dictionary for the data 'resto_id', 'resto_name', and 'cuisine'
resto_new = pd.DataFrame({
    'id': resto_id,
    'resto_name': resto_name,
    'cuisine': resto_cuisine
})
resto_new
```

```python
data = resto_new
data.sample(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
tf = TfidfVectorizer()

# Calculate idf on the 'cuisine' data
tf.fit(data['cuisine'])

# Retrieve the feature names
feature_names = tf.get_feature_names_out()

# Print the feature names
print(feature_names)
```

```python
# Fit and transform the data into a TF-IDF matrix
tfidf_matrix = tf.fit_transform(data['cuisine'])

# Check the size of the TF-IDF matrix
print(tfidf_matrix.shape)
```

```python
# Convert the TF-IDF vectorized matrix into dense matrix format
tfidf_matrix.todense()
```

```python
# Create a DataFrame to visualize the tf-idf matrix
tfidf_df = pd.DataFrame(
    tfidf_matrix.todense(),
    columns=tf.get_feature_names_out(),
    index=data['resto_name']
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# calculate cosine similarity on matrix tf-idf
cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim
```

```python
# Create a DataFrame from the cosine_sim variable with restaurant names as rows and columns
cosine_sim_df = pd.DataFrame(cosine_sim, index=data['resto_name'], columns=data['resto_name'])
print('Shape:', cosine_sim_df.shape)

# Display a sample of the similarity matrix for each restaurant
cosine_sim_df.sample(5, axis=1).sample(10, axis=0)
```

```python
def resto_recommendations(resto_name, similarity_data=cosine_sim_df, items=data[['resto_name', 'cuisine']], k=5):
    """
    Restaurant Recommendations based on similarity dataframe

    Parameters:
    ---
    resto_name : str
                Restaurant name (index of the similarity dataframe)
    similarity_data : pd.DataFrame (object)
                      Symmetric similarity dataframe with restaurants as both index and columns
    items : pd.DataFrame (object)
            Contains both names and other features used to define similarity
    k : int
        Number of recommendations to provide
    ---

    In this function, we take k with the highest similarity values
    on the given index (i).
    """

    # Retrieve data using argpartition to perform an indirect partition along the given axis
    # Convert the dataframe to numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:, resto_name].to_numpy().argpartition(
        range(-1, -k, -1))

    # Retrieve data with the highest similarity from the available index
    closest = similarity_data.columns[index[-1:-(k+2):-1]]

    # Drop resto_name so that the searched restaurant name doesn't appear in the recommendation list
    closest = closest.drop(resto_name, errors='ignore')

    return pd.DataFrame(closest).merge(items).head(k)
```
